```python
import pandas as pd
from sklearn.model_selection import train_test_split
import xgboost as xgb
import numpy as np
```

```python
df = pd.read_excel('mental_health_dataset.xlsx')
df.head()
```

```python
def preprocess_data(df):
    mapping = {
        'Mental_Health_Status': {"Poor": 0, "Fair": 1, "Neutral": 2, "Good": 3, "Excellent": 4},
        'Gender': {"Male": 0, "Female": 1, "Non_Binary": 2},
        'Stress_Life_Changes': {"No": 0, "Yes": 1},
        'Isolation': {"No": 0, "Yes": 1},
        'Sleep_Energy': {"Irritable": 0, "Disrupted": 1, "Normal": 2, "Good": 3},
        'Mood_Emotions': {"Anxious": 0, "Depressed": 1, "Sad": 2, "Neutral": 3, "Happy": 4},
        'Social_Support': {"Isolated": 0, "Limited_support_system": 1, "Strong_support_system": 2},
        'Self_Harm_Thoughts': {"Yes": 0, "Sometimes": 1, "Often": 2, "No": 3},
        'Eating_Habits': {"Disrupted": 0, "Normal": 1, "Healthy": 2},
        'Work_School_Performance': {"Poor": 0, "Fair": 1, "Good": 2, "Excellent": 3}
    }
    
    for column, map_dict in mapping.items():
        df[column] = df[column].map(map_dict)
    
    return df

df = pd.read_excel('mental_health_dataset.xlsx')
df = preprocess_data(df)
df.head()
```

```python
X = df.drop("Mental_Health_Status", axis=1)
Y = df['Mental_Health_Status']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

mymodel = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.01)
mymodel.fit(X_train, Y_train)

mymodel.score(X_test, Y_test)
```

```python
from joblib import dump
dump(mymodel,'./../Savedmodels/ml-models.joblib')
```
